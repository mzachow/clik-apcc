```python
import pandas as pd
```

```python
hindcast_coordinates = pd.read_csv("data/hindcast_coordinate_pairs_filtered.csv", index_col=["lat", "lon"])

hindcast = pd.read_csv("data/preprocessed/pnu.csv", index_col=["lat", "lon"])
hindcast = hindcast.merge(hindcast_coordinates, left_index=True, right_index=True, how="inner")
hindcast = hindcast.reset_index()[["model", "init_year", "init_month", "forecasted_year", "forecasted_month", "lat", "lon", "prec", "t2m"]]
hindcast = (hindcast
            .loc[hindcast["init_month"].between(4,11) & hindcast["forecasted_month"].between(4,11) & hindcast["forecasted_year"].between(1993,2016)]
            .reset_index(drop=True)
            .round(2))
```
